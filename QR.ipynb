## Importing the required module

```python
from MyQR import myqr
```

## Running the QR code generator

The myqr.run() function is used to generate the QR code. It takes two important arguments :

The first argument is the URL or the text that you want to encode into the QR code. In this case, it encodes my GitHub link. The second argument is the file name for the output image.

The code will generate a QR code with the content https://github.com/hoodayashika and save it as a GitHubLink.png

```python
myqr.run('https://github.com/hoodayashika', save_name='yashikahoodaGitHubLink.png')
```

## Displaying the QR code

```python
from PIL import Image
```

```python
# Load and display the QR code image
img = Image.open('yashikahoodaGitHubLink.png')
display(img)
```

# Customization of the QR code

Adding a github backgroud to the QR code

```python
myqr.run(words = 'https://github.com/hoodayashika', picture='picture.png', colorized=True, save_name='my_qr_with_github_background.png')
```

## Displaying the Customized QR code

```python
img_bg = Image.open('my_qr_with_github_background.png')
display(img_bg)
```
